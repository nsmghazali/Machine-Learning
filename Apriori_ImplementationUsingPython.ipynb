```python
import pandas as pd
from apyori import apriori
```

```python
# Load The Dataset

# 'header = None' tells python not to treat the first row as column names
store_data = pd.read_csv('https://raw.githubusercontent.com/yashy1626/ds_dataset/refs/heads/main/store_data.csv', header=None)
store_data.head()
```

```python
store_data.shape
```

```python
# The Apriori library we are going to use requires our dataset to be
# in the form of list of lists, where the whole dataset is a big list
# and each transaction in the dataset is an inner list within the outer big list.
# Currently we have data in the form of a pandas dataframe.
```

```python
transactions = [[item for item in row if pd.notna(item)] for row in store_data.values.tolist()]
transactions
```

```python
# in order to have better results

association_rules = apriori(transactions,
                            min_support=0.0045,
                            min_confidence=0.2,
                            min_lift=3)

association_rules = list(association_rules)
```

```python
for item in association_rules:
  print(item)
```

```python
for item in association_rules:
  pair = item[0]
  items = [x for x in pair]
  print(f'Most Likely To Buy: {items[0]} --> {items[1]}')

  for x in item.ordered_statistics:
    print(f'Confidence: {x.confidence}')
    print(f'Lift: {x.lift}')
    #print(f'Support: {x.support}')
    print('='*50)
```
